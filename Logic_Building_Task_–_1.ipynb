```python
#1. User Login Check
#Problem Statement:Given a username and password, check whether login is successful.
#username = "admin"
#password = "1234"
correct_username = "admin"
correct_password = "1234"

username = input("Enter the username: ")
password = input("Enter the password: ")

if username == correct_username and password == correct_password:
  print("Login Successfull")
else:
  print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer
#Problem Statement:Analyze student results.
#marks = [45, 78, 90, 33, 60]
# Given marks
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Analyze results
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print final result
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
#3. Simple Data Cleaner
#Problem Statement:Clean and standardize user names.
#names = [" Alice ", "bob", " CHARLIE "]
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []

for name in names:
    cleaned_names.append(name.strip().lower())

print("Cleaned Names:", cleaned_names)
```

```python
#4. Message Length Analyzer
#Problem Statement:Analyze message sizes.
#messages = ["Hi", "Welcome to the platform", "OK"]
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    print("Message:", msg)
    print("Length:", len(msg))

    if len(msg) > 10:
        print("Long message detected")

    print()  # blank line for clarity
```

```python
#5. Error Message Detector
#Problem Statement:Detect error messages from system logs.
#logs = ["INFO", "ERROR", "WARNING", "ERROR"]
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR messages:", error_count)
```
